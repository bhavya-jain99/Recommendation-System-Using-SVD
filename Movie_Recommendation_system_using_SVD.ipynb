```
#A = normalised.T #/ np.sqrt(X.shape[0] - 1)
U, S, V = np.linalg.svd(normalised)
```

```
def top_cosine_similarity(data, movieId,top=10):#, top=10
    index = list(mov_list).index(movieId) - 1  # Movie id starts from 1 in the dataset
    movie_row = data[index, :]
    magnitude = np.sqrt(np.einsum('ij, ij -> i', data, data))
    similarity = np.dot(movie_row, data.T) / (magnitude[index] * magnitude)
    #sim=similarity > 0.7
    sort_indexes = np.argsort(-similarity)
    #res = [similarity for similarity in index if similarity >= thres]
    return sort_indexes[:top_n]
    #return res
```

```
def print_similar_movies(movie_data, movieId, mov_list, top_indexes):
    print('Recommendations for {0}: \n'.format(
    movie_data[movie_data.movieId == movieId].title.values,
    movie_data[movie_data.movieId == movieId].genres.values))
    for id in top_indexes:
        print(movie_data[movie_data.movieId == mov_list[id]].title.values,
              movie_data[movie_data.movieId == mov_list[id]].genres.values)
```

```
k = 503
movieId = 94 # (getting an id from movies data)
top_n = 10
sliced = V.T[:, :k] # representative data
indexes = top_cosine_similarity(sliced, movieId, top_n)
```
